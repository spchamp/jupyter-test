```python
import numpy as np
import pandas as pd
# import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt

## example adapted after
## https://learning.oreilly.com/library/view/python-for-data/9781098104023/ch09.html

df = pd.DataFrame(np.random.uniform(size=(6, 4)),
                   index=["Gen X", "Gen Y", "Gen Z", "Gen 0", "Gen Ï€", "Other"],
                   columns=pd.Index(["Blues", "Rock", "Folk", "Classical"], name="Genre"))
sns.set(rc={'figure.figsize':(8.7, 3.80)})
plt.style.use('fivethirtyeight')

b = df.plot.bar(title="Music Ratio")

plt.xticks(rotation=0)
sns.move_legend(b, "upper right", bbox_to_anchor=(1.18, 1))
```
